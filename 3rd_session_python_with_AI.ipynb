## Loops

```python
complex_dict = {
    'student_marks' : {
        'subjects': ['maths, science, english'],
        'marks' : [80, 90, 100, 85, 75] # list is dynamic
    },
    'details': {
        'hobbies' : ['I want to play (badminton)', 'I like reading (books)', 'I watch (movies)'] # list is dynamic
    },
}
```

```python
# calculate percentage for marks using for loops
marks_list = complex_dict['student_marks']['marks']

# percentage = (marks_obtained / total_possible_marks) * 100 
marks_obtained = 0
# len_mark_list = 0
for i in range(0, len(marks_list)):
  print(i)
  # len_mark_list = len_mark_list + 1
  marks_obtained = marks_obtained + marks_list[i]
  

print(marks_obtained)
len_mark_list = len(marks_list)

total_possible_marks = (len_mark_list * 100)

total_possible_marks

percentage = (marks_obtained / total_possible_marks) * 100 
percentage
```

```python
# using list comprehension fetch the hobbies  from hobbies list
hobbies_list = complex_dict['details']['hobbies']
# [for hobies in hobbies_list]
# badminton
[hobbies.split()[-1] for hobbies in hobbies_list]
```

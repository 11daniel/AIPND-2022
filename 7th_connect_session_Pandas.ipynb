## Pandas Series

Pandas Series are very similar to Numpy arrays, except that we can give them named or date time index, instead of just numeric index in Numpy array.

```
# importing pandas and numpy library as pd and np resp(as per convention)
import pandas as pd
import numpy as np
```

```
# python string list
labels = ['a', 'b', 'c', 'd']
# python number list
numbers_list = [10,20,30,40]
# numpy array
np_arr = np.array([10,20,30,45])
# python dict
dic = {'a':10, 'b':20, 'c':30, 'd':40.5}
```

```
# Creating a Series from list
pd.Series(numbers_list)
```

We can see that our list has been converted to pandas series with values and auto index assigned to them. We can provide our own index to Series.

```
# creating series with numbers_list as data and index as labels list
ser = pd.Series(data = numbers_list, index= labels)
ser
```

```
ser['b']
```

We can also create pandas series from other data sources like numpy array and python dictionaries.

```
# pandas series from numpy array
pd.Series(data = np_arr, index = labels)
```

```
# pandas series from dict
pd.Series(dic)
```

Other than numeric values, you can also take create pandas series from string objects.

```
pd.Series(data= labels)
```

We can also add new columns in our existing dataframe. Suppose you want to create a new column by adding values of X and Y columns

```
df['X_Y_SUM'] = df['X'] + df['Y']
```

```
# observing our result, a new column is been created
df
```

To remove a column you can use drop() function. You need to pass name of the column to drop and set the axis value to 1. By default, the axis value is 0, which means it will drop value across row and to drop across columns you need to make axis = 1. There is also a boolean parameter called inplace, whose default value is False. It means that the changes will not be permanent after you drop the value, to avoid any accidental mistakes. To make permanent changes to DataFrame you need to make inplace value True.

```
# Dropping X_Y_SUM column but inplace is False
df.drop('X_Y_SUM', axis=1, inplace=False)
```

```
# checking the df columns, X_Y_SUM is still present
df
```

```
# using drop() function with inplace=True
df.drop('X_Y_SUM',axis=1, inplace=True)
```

```
# checking the df columns, X_Y_SUM is removed now
df
```

We can also use drop() to remove the row. We need to pass the row name. No need to manually change the axis value to 0 as the default value is already 0.

```
df.drop('E')
```

```
# TODO: drop the row E with inplace kept to True
df.drop('E', inplace=True)
```

**Access rows in DataFrame**
Dataframe provides two ways, **iloc** and **loc** to access the rows.
To fetch row values by index name we use **loc**. To fetch row values by index number we use **iloc**.

```
# To get A row, use loc on df
# It returns a series as one row is selected
df.loc['A']
```

**Conditional Selection in DataFrame**

```
# Checking where df is greater than 0
# This will return a boolean dataframe where,
# values are greater than 0.
print(df)
df > 0
```

```
# storing the value in a variable
bool_df = df > 0
```

```
# Now applying to our orignal df
df[bool_df]
```

We can see that we get values where values are greater than 0 and NaN(null values) where the values are not greater than 0. We can complete the above condition in one step too. We generally uses these conditions with rows.

```
# conditional statement in one step
print(df)
df[df>0]
```

```
# series where X column is greater than 0
df['X'] >0
```

```
# getting dataframe values where column X is greater than 0
print(df)
df[df['X'] > 0]
```

In above text we can see that, we get only A and E rows, because the values of X columns is greter than 0 for these rows. See the output of df['X] >0 statement. Only A and E rows are positive there, so we get the datafrmae contaning rows A and E only.

**Multiple Condition Statements**
We have applied one condition statement till now, we can also provide multiple statment on a dataframe. Suppose you want values where X is also greater than 0 and Y is greater than 0.


```
# Here we have surrounded out conditon in (),
# appliend &(and) operator between them
print(df)
df[(df['X']>0) & (df['Y'] >0)]
```

```
# Similarly, you can perform or operation,
# using pipe(|) opeartor
df[(df['X']>0) | (df['Y'] >0)]
```

```
# TODO select values where W and Z columns are greater than 0
df[(df['W'] > 0) & (df['Z']>0)]
```

## Missing Data

In real life dataset there is lot's of missing data. We should know how to deal with this missing data as this is very important and pandas helps us to deal with these types of datasets.

```
# Creating a Dataframe from Dict
# np.nan is used for null or missing values
dic = {'a':[10,20,np.nan],'b':[50,np.nan,np.nan],'c':[10,20,30]}

# creating DataFrame from dict
m_df = pd.DataFrame(dic)
m_df
# Row 0 have no missing value and column c have no missing value
```

**dropna() function**
Pandas provide dropna function which drops the na values. By default pandas will drop any row which have one or more than one missing or nan values

```
# calling dropna on dataframe
# except 0th row all are dropped
m_df.dropna()
```

```
# To make these on columns we can put axis to 1
# a and b columns get dropped due to nan value
m_df.dropna(axis=1)
```

Remember the changes that we had made are not been reflected on the actual DataFrame. To make permanent changes you need to make the value of inplace variable as True

```
# TODO: make changes by making inplace= True
m_df
```

**fillna() function**
fillna() function is used to fill the missing values(nan). You can customize this to fill the nan values with your needs. For example, some columns require nan to be filled with 0 while some may require it to be filled with mean value.

```
# filling the missing value with 0
print(m_df)
m_df.fillna(value = 0)
```

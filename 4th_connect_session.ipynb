### Inheritence

```python
# parent class
class Bird:

  def __init__(self):
      print("Bird is ready")

  def whoisThis(self):
      print("Bird")

  def swim(self):
      print("Swim faster")
```

```python
# child class
class Penguin(Bird):

  def __init__(self):
      # call super() function
      super().__init__()
      print("Penguin is ready")

  def __whoisThis(self):
      super().whoisThis()
      print("Penguin")

  def run(self):
      self.__whoisThis()
      print("Run faster")
```

```python
peggy = Penguin()
```

```python
bird = Bird()
```

```python
peggy.__whoisThis()
```

```python
bird.whoisThis()
```

```python
bird.swim()
```

```python
peggy.swim()
```

```python
peggy.run()
```

```python
bird.run()
```

### Encapsulation

```python
class Computer:
  # protected String name;
  def __init__(self):
      self.__maxprice = 900

  def sell(self):
      print("Selling Price: {}".format(self.__maxprice))

  def setMaxPrice(self, price):
      self.__maxprice = price
```

```python
c = Computer()
c.sell()
```

```python
# c.maxprice = 1500
# c.maxprice
```

```python
# change the price
c.__maxprice = 1000
c.sell()

print(c.__maxprice)
c.sell()
```

```python
# using setter function
c.setMaxPrice(1000)
c.sell()
```

**Numpy Indexing**

Just like python list we can perform indexing operation to select certain values from numpy array. We can use slicing notation(:) to grab the values of certain elements of numpy array.


```
# creating a array from 0 to 9
arr = np.arange(0,10)
arr
```

```
# To select any element of array, we will pass the index that we want
arr[0]
```

```
# using silicing to get values from 1 to 4
# Note that 1 is the starting index which will be included and,
# 5 will be the end index which wil be excluded
arr[0:5]
```

```
# If we want every element upto a certain position
# this will not include 5 index
arr[:5]
```

```
# If we want every element after a certain element
# we want every element after 2nd index
arr[2:]
```

```
# If we wan't to use broadcasting to change a certian values
# Changing first 5 values to 25
arr[:5] = 25
```

```
# Note that the changes are permanent after broadcasting
arr
```

```
# reassigning array
arr = np.arange(0,10)
arr
```

**Conditional Selection**

Numpy provides flexibility to perform conditional selection which is very important feature. Let's see this property with help of a new example

```
arr = np.arange(0,10)
arr
```

```
# suppose we want to know what are elements which are greater than 4
bool_arr = arr > 4
bool_arr
```

In the above example we did broadcasting to check which elements contains value greater than 4. We get an boolean array having True value after 5th index. For rest of the array where value is less than or equal to 4 the value returned False.

```
# Now we can use the bool_arr to get the values which are greater than 4,
# from original array
arr[arr > 4]
```

```
# We can perform the operation directly on arr
arr[(arr > 4) | (arr<8)]
```

## Construct Company A FY2020 Revenue

### Subtask:
Construct a NumPy array for Company A's FY 2020 quarterly revenue using the provided list [20, 40, 40, 50].


**Reasoning**:
To construct the NumPy array for Company A's FY 2020 quarterly revenue, I will first create a Python list with the given values and then convert it into a NumPy array.



```
company_a_fy2020_revenue_list = [20, 40, 40, 50]
company_a_fy2020_revenue = np.array(company_a_fy2020_revenue_list)

print("Company A FY 2020 Revenue List:", company_a_fy2020_revenue_list)
print("Company A FY 2020 Revenue Array:", company_a_fy2020_revenue)
```

## Find Max Revenue Quarter for Company A

### Subtask:
Identify and display the quarter with the maximum revenue for Company A in FY 2020.


**Reasoning**:
To identify the quarter with the maximum revenue, I will use NumPy's `max()` method to find the highest revenue value and `argmax()` to get its index. Since quarters are 1-indexed, I will add 1 to the index to display the correct quarter number.



```
max_revenue_a = company_a_fy2020_revenue.max()
max_revenue_quarter_index_a = company_a_fy2020_revenue.argmax()
max_revenue_quarter_a = max_revenue_quarter_index_a + 1

print(f"Company A FY 2020 Max Revenue: {max_revenue_a}")
print(f"Company A FY 2020 Max Revenue Quarter: Quarter {max_revenue_quarter_a}")
```
